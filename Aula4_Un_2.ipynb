```python
# Aplicando a  aula crie um grafico de um produto hipotetico

import matplotlib.pyplot as plt

#Dados de exemplo
meses = ["janeiro", "fevereiro", "março", "abril", "maio", "junho", "julho"]
vendas = [120, 80, 90, 200, 220, 2200, 50]

# Criar um grafico de barras
plt.bar(meses,vendas,color=['blue', 'green', 'red', 'purple', 'orange', 'brown'])


# Adicionar rotulos aos eixos
plt.xlabel('Mês')
plt.ylabel('Vendas( em unidades)')

# Adicionar um titulo ao grafico
plt.title('Vendas Mensais')

# Adicionar rotulos de dados em cima de cada barra
for i, v in enumerate(vendas):
    plt.text(i, v + 10
            , str(v), ha='center', va='bottom')


# Mostrar o grafico
plt.show()
```
