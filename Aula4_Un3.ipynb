```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='darkgrid') # Opções: Darkgrid, Whitegrid,dark,white,ticks

df_tips = sns.load_dataset('tips')
print(df_tips)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)
# Media por sexo

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])
# O motivo por optar por 'barplot()',muitas vezes se baseia nos parâmetros adicionais e na flexibilidade que ele oferece
# Vamos dar destaque ao parâmetro' Estimator', que por padrão,  calcula a media

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)
# A função 'barplot()' no Seaborn apresenta uma variedade de opções de estatisticas
# Permitindo que os cientistas de dados escolham a metrica que melhor se ajustam aos seus obejetivos
# Por exemplo voce pode calular a soma, a contagem, ou até mesmo outra metrica personalizadas

#I sso é particulamente util quando voçê deseja exibir informações diferentes nas barras
# Como a quantidade(len), Ou soma (sum), em vez da media
```

```python
# Aplicando a aula
# Restaurante que saber qm qual periodo ganha mais gorjetas
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='darkgrid') # Opções: Darkgrid, Whitegrid,dark,white,ticks

df_tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df_tips, estimator=sum, ci=None, palette='Set2')

plt.xlabel('Periodo (Time)')
plt.ylabel('Total de gastos')
plt.title('Total de gastos por periodo( Almoço ou Jantar)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df_tips)#, #estimator=sum, ci=None, palette='Set2')

plt.xlabel('Periodo (Time)')
plt.ylabel('Media de gastos')
plt.title('Media de gastos por periodo( Almoço ou Jantar)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df_tips, palette='Set3')

plt.xlabel('Periodo (Time)')
plt.ylabel('Media de gorjetas')
plt.title('Media de gorjetas por periodo( Almoço ou Jantar)')
plt.show()
```
